# What was the average age of player in the league?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)

nba = pd.read_csv('nbastats2018-2019.csv')

nba.head(10)

# av_age = nba[["Age"]].mean()
# av_age


```

# What player scored the most points?

```python
point_check = nba[["Name", "Points"]]

most_points = point_check.sort_values(by=['Points'], ascending=False).head(50)
most_points
```

# What player had the most blocks during the season? Was it a post player (F/C)?

```python
point_check = nba[["Name", "Blocks"]]

most_points = point_check.sort_values(by=['Blocks'], ascending=False).head(50)
most_points
```

# Based on the regular season, who had the best chance to win a title given their win percentage?

```python
teem = nba.groupby("Team").mean()
teem
```

# What player had the best 3-pt percentage? 

```python
three_point_check = nba[["Name", "FG3%"]]

most_three_points = three_point_check.sort_values(by=['FG3%'], ascending=False).head(50)
most_three_points

It seems like four players have never missed a 3 pointer...
```

# What team had the youngest roster?

```python
yung = nba[["Team", "Age"]]
yungest = yung.groupby("Team")
check = yungest.mean("Age")
checkit = check.sort_values(by=['Age'], ascending=True).head(50)
checkit

Da Bulllllls
```

# Who is the highest paid player during the seasion?

```python
paid = nba[["Name", "Salary"]]

most_paid = paid.sort_values(by=['Salary'], ascending=False).reset_index().head(50)
most_paid
```
